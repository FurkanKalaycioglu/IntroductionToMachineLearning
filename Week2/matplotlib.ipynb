```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.arange(0, 10, 0.23)
print(x)
y = np.sin(x)
print(y)
plt.plot(x, y)
plt.title('Sinus')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python

x= np.random.rand(100)
y= np.random.rand(100)
colors= np.random.rand( 100)
sizes= np.random.randint(100,150,100)
plt.scatter(x, y, c=colors, s=sizes)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data= np.random.randn(1000)
plt.hist(data, bins=20)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Freq')
plt.show()
```

```python
x = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
y = np.random.randint(0, 100, 7)
plt.bar(x, y)
plt.title('Bar plot')
plt.xlabel('Day')
plt.ylabel('Value')
plt.show()
```

```python
data= ['Val1', 'Val2', 'Val3', 'Val4', 'Val5']
values= np.random.randint(0, 100, 5)
plt.pie(values, labels=data)
plt.title('Pie chart')
plt.show()
```

```python
x= np.arange(0, 10, 0.23)
y=np.arange(0, 10, 0.23)
x, y= np.meshgrid(x, y)
r= np.sqrt(x**2 + y**2)
z= np.sin(r)

fig= plt.figure()
ax= fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
plt.title('3D plot')
plt.show()
```
